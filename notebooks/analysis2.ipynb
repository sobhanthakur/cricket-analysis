```python
import pandas as pd
import numpy as np

deliveries = pd.read_csv('../data/deliveries.csv')
matches = pd.read_csv('../data/matches.csv')

deliveries.head(5)
matches.head(5)
```

```python
# Calculate strike rate for all batsmen
runs = deliveries.groupby('batter')['batsman_runs'].sum()
balls = deliveries.groupby('batter')['batsman_runs'].count()
((runs/balls)*100).sort_values(ascending=False)
```

```python
# Top strike rate (minimum 500 balls)
runs = deliveries.groupby('batter')['batsman_runs'].sum()
balls = deliveries.groupby('batter')['batsman_runs'].count()

sr = (runs / balls) * 100

sr[balls >= 500].sort_values(ascending=False)
```

```python
# Lowest strike rate
runs = deliveries.groupby('batter')['batsman_runs'].sum()
balls = deliveries.groupby('batter')['batsman_runs'].count()
((runs/balls)*100).sort_values(ascending=True).head(1)
```

```python
# Economy rate for bowlers
runs = deliveries[~deliveries['extras_type'].isin(['byes', 'legbyes'])] \
        .groupby('bowler')['total_runs'].sum()

balls = deliveries[deliveries['extras_type'] != 'wides'] \
        .groupby('bowler')['ball'].count()

economy = (runs / balls) * 6

economy.sort_values()
```

```python
# Best economy (minimum 300 balls)
runs = deliveries[~deliveries['extras_type'].isin(['byes', 'legbyes'])] \
        .groupby('bowler')['total_runs'].sum()

balls = deliveries[deliveries['extras_type'] != 'wides'] \
        .groupby('bowler')['ball'].count()

economy = (runs / (balls)) * 6
economy[balls >= 300].sort_values()
```

```python
# Worst economy bowlers
runs = deliveries[~deliveries['extras_type'].isin(['byes', 'legbyes'])] \
        .groupby('bowler')['total_runs'].sum()

balls = deliveries[deliveries['extras_type'] != 'wides'] \
        .groupby('bowler')['ball'].count()

economy = (runs / (balls)) * 6
economy.sort_values(ascending=False)
```

```python
# Average runs per wicket
runs = deliveries['total_runs'].sum()
wickets = deliveries[deliveries['is_wicket'] == True]['is_wicket'].count()
runs/wickets
```

```python
# Bowling strike rate
balls = deliveries[deliveries['extras_type'] != 'wides'] \
            .groupby('bowler')['ball'].count()

wickets = deliveries[
    (deliveries['is_wicket'] == 1) &
    (~deliveries['dismissal_kind'].isin(['run out', 'retired hurt', 'obstructing the field']))
].groupby('bowler')['is_wicket'].count()

balls / wickets
```
