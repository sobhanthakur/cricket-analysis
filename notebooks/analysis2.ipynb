```python
import pandas as pd
import numpy as np

deliveries = pd.read_csv('../data/deliveries.csv')
matches = pd.read_csv('../data/matches.csv')

deliveries.head(5)
matches.head(5)
```

```python
# Calculate strike rate for all batsmen
runs = deliveries.groupby('batter')['batsman_runs'].sum()
balls = deliveries.groupby('batter')['batsman_runs'].count()
((runs/balls)*100).sort_values(ascending=False)
```

```python
# Top strike rate (minimum 500 balls)
runs = deliveries.groupby('batter')['batsman_runs'].sum()
balls = deliveries.groupby('batter')['batsman_runs'].count()

sr = (runs / balls) * 100

sr[balls >= 500].sort_values(ascending=False)
```

```python
# Lowest strike rate
runs = deliveries.groupby('batter')['batsman_runs'].sum()
balls = deliveries.groupby('batter')['batsman_runs'].count()
((runs/balls)*100).sort_values(ascending=True).head(1)
```

```python
# Economy rate for bowlers
runs = deliveries[~deliveries['extras_type'].isin(['byes', 'legbyes'])] \
        .groupby('bowler')['total_runs'].sum()

balls = deliveries[deliveries['extras_type'] != 'wides'] \
        .groupby('bowler')['ball'].count()

economy = (runs / balls) * 6

economy.sort_values()
```

```python
# Best economy (minimum 300 balls)
runs = deliveries[~deliveries['extras_type'].isin(['byes', 'legbyes'])] \
        .groupby('bowler')['total_runs'].sum()

balls = deliveries[deliveries['extras_type'] != 'wides'] \
        .groupby('bowler')['ball'].count()

economy = (runs / (balls)) * 6
economy[balls >= 300].sort_values()
```

```python
# Worst economy bowlers
runs = deliveries[~deliveries['extras_type'].isin(['byes', 'legbyes'])] \
        .groupby('bowler')['total_runs'].sum()

balls = deliveries[deliveries['extras_type'] != 'wides'] \
        .groupby('bowler')['ball'].count()

economy = (runs / (balls)) * 6
economy.sort_values(ascending=False)
```

```python
# Average runs per wicket
runs = deliveries['total_runs'].sum()
wickets = deliveries[deliveries['is_wicket'] == True]['is_wicket'].count()
runs/wickets
```

```python
# Bowling strike rate
balls = deliveries[deliveries['extras_type'] != 'wides'] \
            .groupby('bowler')['ball'].count()

wickets = deliveries[
    (deliveries['is_wicket'] == 1) &
    (~deliveries['dismissal_kind'].isin(['run out', 'retired hurt', 'obstructing the field']))
].groupby('bowler')['is_wicket'].count()

balls / wickets
```

```python
# Highest successful chase
matches[matches['result'] == 'wickets']['target_runs'].max()
```

```python
# Lowest defended total
matches[matches['result'] == 'runs']['target_runs'].min()
```

```python
# Highest match total
first = matches['target_runs']
matches['second'] = matches['target_runs'] - matches['result_margin'] - 1
(first+matches['second']).max()
```

```python
# Lowest team score
team_scores = deliveries.groupby(
    ['match_id', 'batting_team']
)['total_runs'].sum().reset_index()

# Merge with matches table
m1 = matches.merge(
    team_scores,
    left_on='id',
    right_on='match_id'
)

# Exclude no result matches
m1 = m1[~m1['result'].isin(['no result'])]

# Lowest team score
m1.sort_values(by='total_runs').head(1)
```

```python
# Team with best chasing record
matches[matches['result'] == 'wickets']['winner'].value_counts().head(1)
```

```python
# Team with best defending record
matches[matches['result'] == 'runs']['winner'].value_counts().head(1)
```

```python
# Highest scoring venues
matches['first'] = matches['target_runs'] -1 
first = matches.groupby('venue')['first'].sum()
second = first - matches.groupby('venue')['result_margin'].sum()
(first+second).sort_values(ascending=False).head(1)
```

```python
# Average first innings score by venue
matches['first'] = matches['target_runs'] -1
matches.groupby('venue')['first'].mean()
```

```python
# Venue with most sixes
runs = deliveries.merge(matches,left_on='match_id',right_on='id')
sixes = runs[runs['batsman_runs'] == 6]
sixes.groupby('venue')['batsman_runs'].sum().sort_values(ascending=False).head(1)
```

```python
# Venue with highest chase
chased_matches = matches[matches['result'] == 'wickets']
chased_matches.groupby('venue')['target_runs'].max().sort_values(ascending=False).head(1)
```

```python
# Venue with highest chase success (in percentage)
venue_matches = matches.groupby('venue')['id'].count()
chased_matches = matches[matches['result'] == 'wickets'].groupby('venue')['target_runs'].count()
((chased_matches/venue_matches)*100).sort_values(ascending=False).head(1)
```

```python
# Most matches played at one stadium
matches['venue'].value_counts().sort_values(ascending=False).head(1)
```

```python
# Best batsman at each venue
venues = deliveries.merge(matches,left_on='match_id',right_on='id')
venues.groupby(['venue','batter'])['batsman_runs'].sum().sort_values(ascending=False)
```
