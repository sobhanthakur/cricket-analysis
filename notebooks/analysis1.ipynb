```python
import pandas as pd
import numpy as np

deliveries = pd.read_csv('../data/deliveries.csv')
matches = pd.read_csv('../data/matches.csv')

deliveries.head(5)
matches.head(5)
```

```python
# Total number of IPL matches played
matches['id'].nunique()
```

```python
# Total number of seasons
matches['season'].nunique()
```

```python
# List all teams that participated
team1 = set(matches['team1'].unique())
team2 = set(matches['team2'].unique())
set(team1.intersection(team2))
```

```python
# Count matches per season
matches.groupby('season')['id'].count()
```

```python
# Count matches played at each venue
matches.groupby('venue')['id'].count()
```

```python
# Which city hosted most matches?
matches['city'].value_counts().idxmax()
```

```python
# Number of matches won by each team
matches['winner'].value_counts()
```

```python
# Number of toss wins by each team
matches['toss_winner'].value_counts()
```

```python
# Number of matches where toss winner also won match
# matches[matches['toss_winner'] == matches['winner']]
# len(matches[matches['toss_winner'] == matches['winner']])
(matches['toss_winner'] == matches['winner']).sum()
```

```python
# Percentage of toss-win → match-win
won = (matches['toss_winner'] == matches['winner']).sum()
total = matches['id'].nunique()
(won/total)*100
```

```python
# Total runs scored in IPL history
deliveries['total_runs'].sum()
```

```python
# Top 10 run scorers
d = deliveries.groupby('batter')['batsman_runs'].sum()
d.sort_values(ascending=False).head(10)
```

```python
# Top 10 six hitters
deliveries[deliveries['batsman_runs'] == 6]['batter'].value_counts().head(10)
```

```python
# Top 10 four hitters
deliveries[deliveries['batsman_runs'] == 4]['batter'].value_counts().head(10)
```

```python
# Players with most balls faced
d = deliveries.groupby('batter')['batsman_runs'].count()
d.sort_values(ascending=False)
```

```python
# Highest individual score
deliveries.groupby(['match_id','batter'])['batsman_runs'].sum().sort_values(ascending=False).head(1)
```

```python
# Total boundaries by each batsman
deliveries[deliveries['batsman_runs'].isin([4, 6])]['batter'].value_counts().sort_values(ascending=False)
```

```python
# Count ducks by batsman
ducks = deliveries.groupby(['match_id','batter'])['batsman_runs'].sum() == 0
ducks.groupby('batter').sum().sort_values(ascending=False)
```

```python
# Most runs in a single season
season_runs = deliveries.merge(matches[['id', 'season']], left_on='match_id', right_on='id')

season_runs.groupby(['season', 'batter'])['batsman_runs'].sum().sort_values(ascending=False).head(1)
```

```python
# Players with most 50s
runs = deliveries.groupby(['batter','match_id'])['batsman_runs'].sum()
((runs >= 50) & (runs < 100)).groupby('batter').sum().sort_values(ascending=False).head(1)
```

```python
# Most wickets taken
# deliveries['dismissal_kind'].unique()
types = ['run out','retired hurt','obstructing the field']
deliveries[deliveries['is_wicket'] == True & ~deliveries['dismissal_kind'].isin(types)]['bowler'].value_counts().head(1)
```

```python
# Most wickets in one season
season_wickets = deliveries.merge(matches[['id', 'season']], left_on='match_id', right_on='id')
season_wickets.groupby('season')['is_wicket'].sum().sort_values(ascending=False).head(1)
```

```python
# Bowlers conceding most runs
valid_run = deliveries[~deliveries['extras_type'].isin(['legbyes','byes'])]
valid_run.groupby('bowler')['total_runs'].sum().sort_values(ascending=False).head(1)
```

```python
# Bowlers with most dot balls
deliveries[deliveries['total_runs'] == 0]['bowler'].value_counts().sort_values(ascending=False).head(1)
```

```python
# Players with most maiden overs
maidens = (deliveries.groupby(['match_id', 'over', 'bowler'])['total_runs'].sum() == 0)
maidens.groupby('bowler').sum().sort_values(ascending=False).head(1)
```
